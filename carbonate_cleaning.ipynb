```python
import pandas as pd
import numpy as np
import os
```

```python
curr_dir = os.getcwd()
df = pd.read_csv(curr_dir + "/CRCP_Carbonate_Chemistry_Atlantic_5f28_f852_5735.csv")

dropped_columns = ["Region", "Mission", "Location", "Survey_design", "Sample_frequency", "accession_url", "SiteID", "UTCDate", "UTCTime", "Date_UTC", "CTDID", "Sample_Depth_m", "Pressure_db"]

df["Year"] = df["Year"].fillna(-1).astype(int)
df["pH_measured"] = df["pH_measured"].fillna(7.000).astype(float)
df["pH_calculated"] = df["pH_calculated"].fillna(7.000).astype(float)

df.rename(columns={"DIC_UMOL_KG": "Dissolved_Inorganic_Carbon_MICRO_MOL_KG", "TALK_UMOL_KG": "Total_Alkalinity_MICRO_MOL_KG", "Aragonite_Sat": "Aragonite_Saturation_State"}, inplace=True)

pd.set_option('display.max_columns', None)

df = df.drop(columns=dropped_columns, axis=1)

df
```
