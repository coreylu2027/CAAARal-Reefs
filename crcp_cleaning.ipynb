```python
import pandas as pd
import numpy as np
import os
```

```python
curr_dir = os.getcwd()

df = pd.read_csv(curr_dir + '/CRCP_BMU_Erosion_Data_Coral_Sites_Atlantic_d06f_af0d_95aa.csv')

pd.set_option('display.max_columns', None)

dropped_columns = ['accession_url', 'Deployment', 'Recovery', 'RegionCode', 'InstrumentType', 'Fate', 'BlockDimensions', 'SubRegion', 'SiteName','Location', 'ProjectCode', 'Depth']

df = df.drop(dropped_columns, axis=1)

df = df.rename(columns={"tag_number": "ID"})

df = df.dropna(subset=['ID'])

df['ID'] = df["ID"].astype(int)

df.insert(1, "year", df["time"].str[:4])

df['PreWeightBlock'] = df['PreWeightBlock'].fillna(-1).astype(float)
df['PreVolume'] = df['PreVolume'].fillna(-1).astype(float)
df['PostCTGrazing'] = df['PostCTGrazing'].fillna(-1).astype(float)
df['PostCTGrazingRate'] = df['PostCTGrazingRate'].fillna(-1).astype(float)
df

df.to_csv('BMU_Erosion_Data_Cleaned.csv')
```
