```python
import pandas as pd
```

```python
# Define the filter criteria
target_year = 2000
chunk_size = 50000  # Adjust based on your available memory

# Create an empty list to store filtered chunks
filtered_chunks = []

cols_to_keep = ['SID', 'SEASON', 'NAME', 'ISO_TIME', 'LAT', 'LON', 'WMO_WIND', 'WMO_PRES', 'USA_WIND', 'USA_PRES', 'USA_SSHS', 'STORM_SPEED', 'STORM_DIR', 'DIST2LAND', 'LANDFALL', 'TRACK_TYPE']

# Read and filter in one pass using chunks
for chunk in pd.read_csv(
        "../../data/raw_large/climateDataibtrac.csv",
    na_values=["nd", "ND", "Nd"],
    low_memory=False,
    chunksize=chunk_size,
    usecols=cols_to_keep
):
    # Convert SEASON to numeric within the chunk
    chunk['SEASON'] = pd.to_numeric(chunk['SEASON'], errors='coerce')

    # Filter the chunk
    filtered_chunk = chunk[chunk['SEASON'] >= target_year]

    # Append the result
    filtered_chunks.append(filtered_chunk)

# Combine filtered chunks into a single DataFrame
df = pd.concat(filtered_chunks, ignore_index=True)

df
```

```python
df.to_csv("../../data/processed/IBTrACS/filtered_cyclone.csv", index=False)
```
