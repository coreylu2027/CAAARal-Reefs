```python
import pandas as pd
import os
```

```python
curr_dir = os.getcwd()
parent_dir = os.path.join(curr_dir, "..", "..")
parent_dir_abs = os.path.abspath(parent_dir)

df = pd.read_csv(parent_dir_abs + "/data/raw/NOAA/global_bleaching_environmental.csv")

pd.set_option('display.max_columns', None)

dropped_cols = ['Site_ID', 'Sample_ID', 'Data_Source', 'Reef_ID', 'Realm_Name', 'Ecoregion_Name', 'Country_Name', 'State_Island_Province_Name', 'City_Town_Name', 'Site_Name', 'Date_Day', 'Date_Month',
                'Date_Year', 'Substrate_Name', 'Windspeed']

df = df.drop(columns=dropped_cols, axis=1)

df.to_csv('NOAA_Global_Bleaching_Cleaned.csv')
```
